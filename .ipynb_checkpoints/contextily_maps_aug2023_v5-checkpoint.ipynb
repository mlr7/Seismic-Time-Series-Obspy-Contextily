### Imports

```python
%matplotlib inline

import contextily as ctx
import geopandas
import rasterio
from rasterio.plot import show as rioshow
from shapely.geometry import box
import matplotlib.pyplot as plt
```

### Define lat/lon box

```python
# ridgecrest
#west = -118.016
#south = 35.3053
#east =  -117.0115
#north = 36.137

# socal
#west = -120.5 #-121.000
#south = 32.00
#east =  -114.000
#north = 35.00 # 37.00

# ventura county (match obspy data pull)
west = -120
south = 33
east =  -118
north = 35
```

### Get map

```python
img = []
ext = []

img, ext = ctx.bounds2raster(west,
                             south,
                             east,
                             north,
                             #"world_watercolor.tif",
                             #"ridgcrest2019b.tif",
                             #"test2.tif",
                             #"socal2.tif",
                             "ventura_county1.tif",
                             #source=ctx.providers.Stamen.Watercolor,
                             source=ctx.providers.OpenStreetMap.Mapnik,
                             ll=True
                            )
```
