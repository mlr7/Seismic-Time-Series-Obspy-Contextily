### Get seismic data

```python
iris_client = Client("IRIS", timeout=600)
```

```python
#cat = iris_client.get_events(minlat= 33,# 32,
#                           maxlat= 35 ,#36, #35,
#                           minlon=-120, #-121, #-119,
#                           maxlon=-118, #-117,
#                           starttime=UTCDateTime("2023-08-19"),
#                           #endtime=UTCDateTime("2023-08-21"),
#                           endtime=UTCDateTime("2023-08-23"),
#                           #starttime=UTCDateTime("2019-01-01"),
#                           #endtime=UTCDateTime("2023-02-28"),
#                           minmag=0.0) # 2.0
#cat
```

```python
ventura_xmax = -118.632494
ventura_xmin = -119.636302

ventura_ymax = 34.901171
ventura_ymin = 33.163493

ventura_maxlon = math.ceil(ventura_xmax)
ventura_minlon = math.floor(ventura_xmin)

ventura_maxlat = math.ceil(ventura_ymax)
ventura_minlat = math.floor(ventura_ymin)

print(ventura_xmax, ventura_maxlon)
print(ventura_xmin, ventura_minlon)
print(" ")
print(ventura_ymax, ventura_maxlat)
print(ventura_ymin, ventura_minlat)
```

```python
cat_ventura = []
cat_ventura = iris_client.get_events(
                           minlat= ventura_minlat,#33,# 32,
                           maxlat= ventura_maxlat,#35 ,#36, #35,
                           minlon= ventura_minlon, #-120, #-121, #-119,
                           maxlon= ventura_maxlon, #-118, #-117,
                           starttime=UTCDateTime("2023-08-17"),
                           #starttime=UTCDateTime("2023-08-19"),
                           #endtime=UTCDateTime("2023-08-21"),
                           endtime=UTCDateTime("2023-08-23"),
                           #starttime=UTCDateTime("2019-01-01"),
                           #endtime=UTCDateTime("2023-02-28"),
                           minmag=0.0) # 2.0

cat_ventura
```

```python
cat = cat_ventura
```

```python
times = []
lats = []
lons = []
deps = []
magnitudes = []
magnitudestype = []
for event in cat:
    if len(event.origins) != 0 and len(event.magnitudes) != 0:
        times.append(event.origins[0].time.datetime)
        lats.append(event.origins[0].latitude)
        lons.append(event.origins[0].longitude)
        deps.append(event.origins[0].depth)
        magnitudes.append(event.magnitudes[0].mag)
        magnitudestype.append(event.magnitudes[0].magnitude_type )
```

### Convert raw data to pandas dataframe

```python
#df = pd.DataFrame({'lat':lats,'lon':lons,'depth':deps,
#                   'mag':magnitudes,'type':magnitudestype}, 
#                  index = times)

df = pd.DataFrame({'datetime':times, 'lat':lats,'lon':lons,'depth':deps,
                   'mag':magnitudes,'type':magnitudestype})

print(df.shape)
print(df.columns)
display(df.head(4))
display(df.tail(4))
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
df = df.sort_values(by='datetime')
df = df.reset_index(drop=True)
display(df.head(3))
display(df.tail(3))
```

```python
print(np.min(df['mag']))
print(np.max(df['mag']))
```
